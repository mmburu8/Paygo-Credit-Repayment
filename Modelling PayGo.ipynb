## ANOMALY DETECTION: ISOLATION FOREST

```python
#IDENTIFYING OUTLIERS IN DATASET
X_train = X_train.values
iso = IsolationForest(contamination= 0.1)
xhat = iso.fit_predict(X_train)
```

```python
# select all rwos that are not outliers
mask = xhat != -1
X_train, y_train = X_train[mask, :], y_train[mask]
```

HYPERPARAMETER TUNING NEURAL NETWORK

```python
# make model
def create_model(init_mode= "normal"):
    model = Sequential()
    model.add(Dense(170, input_dim= 268, activation= "relu"))
    model.add(Dense(100, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(90, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(80, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(72, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(68, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(62, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(57, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(52, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(47, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(41, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(37, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(26, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(16, activation= "relu", kernel_initializer= init_mode))
    model.add(Dense(6))
    model.compile(optimizer= "RMSprop", loss= "mean_squared_error", )
    return model
```

```python
nn_model= KerasRegressor(build_fn=create_model, epochs= 150, batch_size= 8, verbose= 0)
```

```python
# hyperparameter tuning optimizer
# makin dictionary
optimizer = ["SGD", "RMSprop", "Adam", "Adamax", "Nadam"]
param_random = dict(optimizer= optimizer)
grid_random = GridSearchCV(estimator= nn_model, param_grid=param_random, n_jobs= -1, cv= 3, scoring= "neg_mean_squared_error")
random_result = grid_random.fit(X_train, y_train)
print(random_result.best_score_)
print(random_result.best_params_)
```

```python
# hyperparameter tuning batch_size and epochs
param_test = {
    "batch_size": [int(x) for x in np.linspace(start= 8, stop= 50, num= 6)],
    "epochs": [int(x) for x in np.linspace(start= 50, stop= 500, num= 10)]
}
randoms = RandomizedSearchCV(nn_model, param_test, n_jobs= -1, cv= 3, scoring= "neg_mean_squared_error")
randoming = randoms.fit(X_train, y_train)
print(randoming.best_score_)
print(randoming.best_params_)
```

```python
# hyperparameter tuning weight initialization
init_mode= ["uniform", "lecun_uniform", "normal", "zero", "glorot_normal", "glorot_uniform", "he_normal", "he_uniform"]
param_grid = dict(init_mode= init_mode)
grid = GridSearchCV(estimator= nn_model, param_grid= param_grid, n_jobs= -1, cv= 3, scoring= "neg_mean_squared_error")
grid_result = grid.fit(X_train, y_train)
print(grid_result.best_score_)
print(grid_result.best_params_)
```
